```polyglot-notebook
using System;                          // Core system types and functionality
using System.ComponentModel;           // For Description attributes on agent tool methods
using System.ClientModel;             // Client model abstractions for API interactions

using Microsoft.Extensions.AI;        // Core AI service abstractions and interface
using Microsoft.Agents.AI;           // Agent framework classes and orchestration system
using OpenAI;                         // OpenAI client library for GitHub Models integration
```

```polyglot-notebook
 using DotNetEnv;
```

```polyglot-notebook
Env.Load("../../../.env");
```

```polyglot-notebook
var foundrylocal_endpoint = Environment.GetEnvironmentVariable("FOUNDRYLOCAL_ENDPOINT") ?? throw new InvalidOperationException("FoundryLocal_Endpoint is not set.");
var foundrylocal_model_id = Environment.GetEnvironmentVariable("FOUNDRYLOCAL_MODEL_DEPLOYMENT_NAME") ?? "Qwen3-0.6b-cpu";
```

```polyglot-notebook
var openAIOptions = new OpenAIClientOptions()
{
    Endpoint= new Uri(foundrylocal_endpoint)
};
```

```polyglot-notebook
foundrylocal_endpoint
```

```polyglot-notebook
foundrylocal_model_id
```

```polyglot-notebook
AIAgent agent = new OpenAIClient(new ApiKeyCredential("nokey"), openAIOptions).GetChatClient(foundrylocal_model_id).AsIChatClient().AsAIAgent(
    instructions:"You are a helpful assistant.");
```

```polyglot-notebook

Console.WriteLine(await agent.RunAsync("Can you introduce yourself?"));
```

```polyglot-notebook
await foreach (var update in agent.RunStreamingAsync("Can you introduce yourself?"))
{
    Console.Write(update);
}
```
